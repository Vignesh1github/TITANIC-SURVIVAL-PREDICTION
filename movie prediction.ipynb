```python
import pandas as pd

# Sample DataFrame
data = {'Votes': ['$5.16M', '$1.25K', '$123']}
X_train = pd.DataFrame(data)

# Function to convert values to float
def clean_currency(value):
    if isinstance(value, str):
        value = value.replace('$', '').replace(',', '')  # Remove dollar sign and commas
        if 'M' in value:
            value = float(value.replace('M', '')) * 1_000_000  # Convert millions to numeric
        elif 'K' in value:
            value = float(value.replace('K', '')) * 1_000  # Convert thousands to numeric
        else:
            value = float(value)
    return value

# Apply cleaning function to the 'Votes' column
X_train['Votes'] = X_train['Votes'].apply(clean_currency)

# Verify the changes
print(X_train)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Sample DataFrame with missing target values
data = {'Feature1': [1, 2, 3, 4, 5], 'Feature2': [10, 20, 30, 40, 50], 'Target': [1, 2, None, 4, None]}
df = pd.DataFrame(data)

# Split data
X = df[['Feature1', 'Feature2']]
y = df['Target']

# Remove rows with NaNs in target variable
valid_indices = y.notna()
X_clean = X[valid_indices]
y_clean = y[valid_indices]

# Split clean data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=0.2, random_state=42)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize and train model
model = RandomForestClassifier()
model.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test_scaled)

# Output the predictions
print("Predictions on the test set:", y_pred)

# Compare with actual values
print("Actual values:", y_test.values)

# Evaluate the model performance (accuracy)
test_score = model.score(X_test_scaled, y_test)
print("Test accuracy:", test_score)
```
